# Understanding the Statistical Summary of the data

```python
dfo.describe()
```

```python
df2 = pd.get_dummies(dfo["Gender"])
df3 = pd.concat((df2,dfo), axis=1)
df3 = df3.drop(["Gender"], axis=1)
df3 = df3.drop(["Male"], axis=1)
result = df3.rename(columns={"Female": "Gender"})
dfc = result
dfc
```

# Drawing Inferences/Insights from the data

'Age', 'intercanine distance casts', 'right canine width casts', 'right canine index intra oral' are the independent variables.

# Model Building

```python
feature_cols = ['Age','intercanine distance casts', 'right canine width casts', 'right canine index intra oral']
X = dfc[feature_cols]
Y = dfc.Gender
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size= 0.3, random_state = 40)
```

# Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
clf = LogisticRegression(random_state=40)  
abc =AdaBoostClassifier(n_estimators=50, base_estimator=clf,learning_rate=1)
abc.fit(X_train, Y_train)
Y_pred = abc.predict(X_train)
Y_pred1 = abc.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred1))
print("Precision:",metrics.precision_score(Y_test, Y_pred1))
print("Recall:",metrics.recall_score(Y_test, Y_pred1))
print("F1-Score:",metrics.f1_score(Y_test, Y_pred1))
```

```python
print(metrics.classification_report(Y_train, Y_pred))
print(metrics.confusion_matrix(Y_test, Y_pred1))
```

```python
print(cross_val_score(clf,X_train,Y_train,cv=30,scoring="accuracy").mean())
print(cross_val_score(clf,X_test,Y_test,cv=30,scoring="accuracy").mean())
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=40)
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_train)
Y_pred1 = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred1))
print("Precision:",metrics.precision_score(Y_test, Y_pred1))
print("Recall:",metrics.recall_score(Y_test, Y_pred1))
print("F1-Score:",metrics.f1_score(Y_test, Y_pred1))
```

```python
print(metrics.classification_report(Y_train, Y_pred))
print(metrics.confusion_matrix(Y_test, Y_pred1))
```

```python
print(cross_val_score(clf,X_train,Y_train,cv=30,scoring="accuracy").mean())
print(cross_val_score(clf,X_test,Y_test,cv=30,scoring="accuracy").mean())
```

# Support Vector Classifier
